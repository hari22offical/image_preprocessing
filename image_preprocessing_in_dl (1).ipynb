```python
# importing the libraries
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

```python
# loading an image through matplotlib.image module
#converting the image into numpy array
img=mpimg.imread('/content/dog.jpg')
```

```python
#convert numpy array into image
img_plot=plt.imshow(img)
plt.show()
```

Resizing the image using Pillow library

```python
from PIL import Image
img=Image.open('/content/dog.jpg')
img_reshape=img.resize((150,200))
```

```python
img_reshape.save('dog_reshape.jpg')
```

```python
img=mpimg.imread('/content/dog_reshape.jpg')
img_plot=plt.imshow(img)
plt.show()
```

```python
img.shape
```

**Converting RGB images to Grayscale image using OpenCV**

```python
#importing open cv libraries
import cv2
img=cv2.imread('/content/dog.jpg')
```

```python
type(img)
```

```python
g_img=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
```

```python
print(g_img.shape)
```
